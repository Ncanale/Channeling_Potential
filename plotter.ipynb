```python
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib widget
# plt.rcParams['text.usetex'] = True

# PWO_X0 = 9 ## mm 
PWO_X0 = 0.9 ## cm 
def plotter (file_name):
    df = pd.read_csv(file_name, sep=" ", header=None, names=['id', 'X_0', 'NaN', 'Signal'])
    dx =[df["X_0"][0]]+ [df["X_0"][i+1] - df["X_0"][i] for i in range(len(df["X_0"])-1)]
    adx = sum(dx)/len(dx)
    # print("ave dx =",adx )

    Y_plot = [df["Signal"][i]*(PWO_X0/adx)/sum(df["Signal"])*100 for i in range(len(df["Signal"]))] 
    # X_plot = [df["X_0"][i]/PWO_X0 for i in range(len(df["X_0"])) ] #sono gi normalizzate
    return df["X_0"], Y_plot
    # return X_plot, Y_plot


file_list = ["Energy_deposit_20GeV_mod_gamma.txt", "Energy_deposit_20GeV_mod.txt",
            "Energy_deposit_20GeV_standard_gamma.txt","Energy_deposit_20GeV_standard.txt",
            "Energy_deposit_120GeV_mod_gamma.txt", "Energy_deposit_120GeV_mod.txt",
            "Energy_deposit_120GeV_standard_gamma.txt","Energy_deposit_120GeV_standard.txt"]

# print (file_list[0].split("Energy_deposit_")[1].split(".txt")[0])
linestile = ["-",":"]
plt.title("")
# plt.plot(plotter("Energy_deposit_120GeV_mod_gamma.txt")[0],plotter("Energy_deposit_120GeV_mod_gamma.txt")[1])
for i in range (0,len(file_list)):
    label = file_list[i].split("Energy_deposit_")[1].split(".txt")[0]
    # print (label)
    # plt.plot(plotter(file_list[i])[0],plotter(file_list[i])[1], label = label,linestyle = linestile[i%2])
    plt.plot(plotter(file_list[i])[0],plotter(file_list[i])[1], label = label,linestyle = linestile[i%2])
plt.legend()
plt.ylabel('Fraction of primary energy deposit \n per radiation length [ $\% / X_0$ ]')
plt.xlabel('depth [ $X_0$]')
plt.grid()
plt.show()
```
